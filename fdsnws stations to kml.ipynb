```python
%matplotlib notebook 
from obspy.clients.fdsn import Client
from obspy import UTCDateTime
```

```python
client = Client('http://165.98.224.52:8081/')
starttime = UTCDateTime()-60*60*24*360
endtime = UTCDateTime()
inventory = client.get_stations(network="*", station="*",
                                    starttime=starttime,
                                    endtime=endtime, 
                                    lat=11.984562,
                                    lon=-86.168006,
                                    minradius=0.0,
                                    maxradius=0.1,
                                    level="CHAN")
inventory.write("sc-local-proc_fdsnws_station_masaya_1year.kml", format="KML")
inventory.plot(color_per_network=True,projection="local")
```
