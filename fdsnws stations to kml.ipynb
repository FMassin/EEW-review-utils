```python
%matplotlib notebook 
from obspy.clients.fdsn import Client
from obspy import UTCDateTime
```

```python
client = Client('http://165.98.224.52:8081/')
starttime = UTCDateTime()-60*60*24*360
endtime = UTCDateTime()
inventory = client.get_stations(network="*", station="*",
                                    starttime=starttime,
                                    endtime=endtime, level="CHAN")
```

```python
inventory.write("sc-local-proc_fdsnws_station_1year.kml", format="KML")
```

```python
inventory.plot(projection="local", color_per_network=True)  
```
