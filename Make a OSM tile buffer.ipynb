```python
import urllib3
import os


url ="http://services.arcgisonline.com/arcgis/rest/services/NatGeo_World_Map/MapServer/tile"
path = "/Users/fmassin/Documents/osmtiles/"
http = urllib3.PoolManager()

for l in range(0,6):
    for y in range(0,2**l):
        for x in range(0,2**l):
            tile = url+"/"+str(l)+"/"+str(y)+"/"+str(x)
            directory = path+"/"+str(l)+"/"+str(x)
            file = directory+"/"+str(y)+".jpg"
            
            r = http.request('GET', tile, preload_content=False)

            flag = 0
            os.makedirs( directory, exist_ok=True );
            with open(file, 'wb') as out:
                while True:
                    data = r.read()
                    if not data:
                        break
                        
                    out.write(data)
                    flag = 1
            r.release_conn()
```
